```python
def ver_elementos(obj, todo=True):
  enum=1
  si_color='\033[0;30m'; no_color='\033[0m'
  # Coloreado Purpura \033[0;95m
  print(si_color + 'TIPO: ' + no_color + str(type(obj)) + '\n')
  print(si_color + 'DOC: ' + no_color + type(obj).__doc__ + '\n')
  for elemento in dir(obj):
    if todo == False and '_' in elemento:
      pass
    else:
      ver='obj.' + elemento
      # Coloreado Rojo \033[0;91m
      si_color='\033[0;30m'
      print(si_color + 'Elemento ' + str(enum) + ': ' + no_color + ver + '\n')
      if 'method' in str(eval(ver)):
        # Coloreado Amarillo \033[0;95m
        si_color='\033[0;95m'
        descripcion='Metodo'
      elif 'function' in str(eval(ver)):
        # Coloreado Azul \033[0;94m
        si_color='\033[0;94m'
        descripcion='Funcion'
      elif 'class' in str(eval(ver)):
        # Coloreado Cyan \033[0;96m
        si_color='\033[0;96m'
        descripcion='Clase '
      else:
        # Coloreado Verde \033[0;92m
        si_color='\033[0;92m'
        descripcion='Atribu'
      print('\t' + si_color + descripcion + ' --> ' + no_color + str(eval(ver)) )
      print('\t' + si_color + 'Tipo   --> ' + no_color + str(type(eval(ver))) )
      print('\t' + si_color + 'Doc    --> ' + no_color + str(eval(ver).__doc__) + '\n')
      enum=enum+1

# Entero
var1=1
print("\n-------------- SHORT INFO ABOUT OBJECT --------------\n")
print(f"Object --> {var1}, type --> {type(var1)}")

obj_attributes = [attr for attr in dir(var1) ]	
print(f"Number of attributes -> {len(obj_attributes)}")
print('Attributes ->' ,obj_attributes, '\n')

# ----------------------------------------------

print("\n-------------------- more details -------------------\n")
ver_elementos(var1,True)
```
