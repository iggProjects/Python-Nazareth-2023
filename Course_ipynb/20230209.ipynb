```python
# Cálculo del precio con iva
iva = "25" #@param ["10", "20", "25"]
iva =float(iva)
precio_sin_iva = 200 #@param {type:"number"}
print(f"\t\tprecio sin iva --> {precio_sin_iva:.2f}\n")
print(f"\t\tprecio con iva del {iva} % --> {(precio_sin_iva * (100 + iva)/100):.2f}\n")
print(f"\t\ttype of result --> {type(precio_sin_iva * (100 + iva)/100)}\n")
```

```python
# Conversión temperatura fharenheit a celsius 
grados_F = 32 #@param {type:"number"}
grados_F = float(grados_F)
Convers_grad_C = (grados_F - 32) * 0.5556
# Ejemplo (75F - 32) x 0.5556
print(f"Temperatura en grados celsius --> {Convers_grad_C} | data type {type(Convers_grad_C)}")
```

```python
# Conversión temperatura celsius a fHarenheit
grados_C = 0 #@param {type:"number"}
grados_C = float(grados_C)
Convers_grad_F = (grados_C * 1.8 ) + 32
# Ejemplo (50C x 1.8) + 32 = 122F
print(f"Temperatura en grados fharenheit --> {Convers_grad_F} | data type {type(Convers_grad_F)}")
```
