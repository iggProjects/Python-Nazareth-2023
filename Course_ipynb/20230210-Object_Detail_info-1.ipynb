```python
def ver_objetos(obj):
	enum=1
	si_color='\033[0;91m'
	no_color='\033[0m'
	print(si_color + 'TIPO:' + no_color, str(type(obj)) , '\n')
	for elemento in dir(obj):
		ver='obj.' + elemento
		print(str(si_color + 'Elemento: ' + str(enum) + ' ' + no_color + ver))  
		if 'method' in str(eval(ver)):
			print('\tMetodo --> ', eval(ver)) 
		else:
			print('\tAtribu --> ', eval(ver))
			print('\tTipo --> ', type(eval(ver)), '\n')
		enum=enum+1    

si_color='\033[0;91m'
no_color='\033[0m'

# Entero
var1=1

print(f"Object {var1}, type {type(var1)}")

print("\n--------------attributes of object-------------------\n")
obj_attributes = [attr for attr in dir(var1) 
	if not attr.startswith('__')]
print('\tAttributes ->' ,obj_attributes, '\n')

# ----------------------------------------------

print("\n----------------- more details ----------------\n")
ver_objetos(var1)
```
