```python
"""
Lists and Tuples

differences between lists and tuples

https://www.geeksforgeeks.org/python-difference-between-list-and-tuple/
https://builtin.com/software-engineering-perspectives/python-tuples-vs-lists

"""

# lists EXAMPLE

# CONSTANTS
color_Y='\033[0;94m'
color_N='\033[0m'

compras = ['platanos', 'manzanas', 'leche']

print(color_Y)

# Añadir “galletas” y “zumo” a la lista de compras
compras=compras + ['galletas','zumo']

# Mostrar todo la lista
print(f"List: {compras}\n")

# Mostrar el segundo y tercer elemento
print(f"second and third products: {compras[1]}, {compras[2]}\n")

# Mostrar los últimos 2 elementos - “galletas” y “zumo”
leng=len(compras)
print(f"last two products: {compras[leng-2]}, {compras[leng-1]}\n")

# Cambiar “zumo” por “zumo de naranja"
# https://www.geeksforgeeks.org/how-to-replace-values-in-a-list-in-python/
# finding index of 'zumo'
i = compras.index('zumo')
compras[i]='zumo de naranja'
print(f"'zumo' item replaced -- list updated: {compras}\n")

# Quitar la última compra que has añadido a la lista
compras.pop()
print(f"Last iten removed -- list updated: {compras}\n")

# Insertar “limones” en una posición del medio de la lista
import math
midpoint=math.floor(len(compras)/2)
compras=compras[0:midpoint] + ['limones'] + compras[midpoint:] 
print(f"'limones' inserted at middle position -- list updated: {compras}\n")

# Mostrar la lista en orden alfabético
# https://docs.python.org/3/howto/sorting.html
print(f"list updated sort in alphabetic order: {sorted(compras)}\n")

print(color_N)
```
